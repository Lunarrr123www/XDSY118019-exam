## Project
Collatz Conjecture:Given an integer  n , if it is even, then divide it by 2, if it is odd, then multiply it by 3 and then add 1. Conjecture: repeat this process will eventually lead to the number 1.    
Task:
1. Given a positive integer, print out its Collatz sequence.
2. For positive integers n=3,6,9,12,15,18,21,24,27,30, plot the convergence of their Collatz sequences with the ordinal number of their koraz sequence as the horizontal axis and the sequence value as the vertical axis.

### The explaination to the following code
Put a number into the following code and run it, then you will get its Collatz sequence. Take "27" for example.

------
#### The problem-solving ideas
First, write a loop to calculate the results of the Collatz Conjecture, then crate a blank list "b" and add every output to the list, finally, print out the list to show the given number's Collatz sequence.

```python
a = int(input())
b = []
while a != 1:
    if a % 2 == 0:        # If a is even
        a = a / 2
        b.append(a)
    else:                 # If a is odd
        a = a * 3 + 1
        b.append(a)
print(b)
```

#### The problem-solving ideas
Create a function to calculate the given numbers' Collatz sequence and plot them by using the Pyplot module.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
def check(num):                                 # Create a function to calculate the Collatz sequence.
    times = 0
    numbers = []
    while True:
        times = times + 1
        if num % 2 == 0:
            num = num // 2
        else:
            num = num*3 + 1
        numbers.append(num)
        if num == 1:
            break
    return range(times), numbers    

for _ in range(1):
    for num in [3,6,9,12,15,18,21,24,27,30]:    # Plot the Collatz sequences of the given numbers.
        plt.plot(*check(num), label=str(num))
plt.legend()
plt.show()
```
